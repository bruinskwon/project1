## Plot and Visualization

### **The Dataset**
Food servers’ tips in restaurants may be influenced by many
factors, including the nature of the restaurant, size of the party, and table
locations in the restaurant. Restaurant managers need to know which factors
matter when they assign tables to food servers. For the sake of staff morale,
they usually want to avoid either the substance or the appearance of unfair
treatment of the servers, for whom tips (at least in restaurants in the United
States) are a major component of pay.
In one restaurant, a food server recorded the following data on all cus-
tomers they served during an interval of two and a half months in early 1990.
The restaurant, located in a suburban shopping mall, was part of a national
chain and served a varied menu. In observance of local law, the restaurant
offered to seat in a non-smoking section to patrons who requested it. Each
record includes a day and time, and taken together, they show the server’s
work schedule.
([Reference](https://www.kaggle.com/ranjeetjain3/seaborn-tips-dataset))


The dataset, `tips`' schema is as follows:

* `total_bill`	
* `tip`	
* `sex`
* `smoker`	
* `day`	
* `time`	
* `size`

```
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

tips = sns.load_dataset("tips")
tips.head()
```

How to check the shape of `tips` dataset?

```
tips.shape
```

Count plot of day

```
sns.countplot(tips.day)
plt.show()
```

Change color

```
sns.countplot(tips.day, 
              palette="cubehelix")
plt.show()
```

```
sns.countplot(tips.day, 
              palette=["#9b59b6", "#3498db", "#95a5a6", "#e74c3c"])
plt.show()
```

[Color Codes](https://python-graph-gallery.com/100-calling-a-color-with-seaborn//)

```
sns.countplot(tips.day,
              palette=["b", "r", "sandybrown", "skyblue"])
plt.show()
```

Barplot with tip by day

```
sns.barplot(x='day', y='tip', ci=None, data=tips)
plt.show()
```

```
sns.set(rc={'figure.figsize':(12,8)})
sns.barplot(x='day', y='tip', ci=None, data=tips)
plt.show()
```

```
sns.reset_orig()
sns.barplot(x='day', y='tip', ci=None, data=tips)
plt.show()
```

```
plt.figure(figsize=(15,8))
sns.barplot(x='day', y='tip', ci=None, data=tips)
plt.show()
```

```
sns.barplot(x='day', y='tip', ci=None, data=tips)
plt.show()
```

```
sns.set_style('darkgrid')
sns.set_palette("RdBu")
sns.set_context("poster") # talk, paper, notebook

sns.barplot(x='day', y='tip', ci=None, data=tips)
plt.show()
```

```
sns.reset_orig()
sns.barplot(x='day', y='tip', ci=None, data=tips)
plt.show()
```

```
sns.scatterplot(x='total_bill', y='tip', data=tips)
plt.show()
```

```
sns.scatterplot(x='total_bill', y='tip', data=tips,
                hue="smoker")
plt.show()
```

```
sns.scatterplot(x='total_bill', y='tip', data=tips,
                hue="smoker", hue_order=['Yes', 'No'])
plt.show()
```

```
plt.figure(figsize=(20,10))
sns.scatterplot(x='total_bill', y='tip', data=tips,
                hue="smoker", hue_order=['Yes', 'No'],
                size='day')
plt.show()
```

```
plt.figure(figsize=(20,10))
sns.scatterplot(x='total_bill', y='tip', data=tips,
                hue="smoker", hue_order=['Yes', 'No'],
                size='day',
                style='sex')
plt.show()
```

```
sns.relplot(x='total_bill', y='tip', data=tips)
plt.show()
```

```
sns.relplot(x='total_bill', y='tip', data=tips, 
            kind='scatter',
            col='smoker')
plt.show()
```

```
sns.relplot(x='total_bill', y='tip', data=tips, 
            kind='scatter',
            col='smoker',
            row='time')
plt.show()
```

## Dataset 2: Real World Dataset

```
path = 'https://raw.githubusercontent.com/RamiKrispin/coronavirus/master/csv/coronavirus.csv'

df = pd.read_csv(path)
df.head()
```

```
df.shape
```

```
df.info()
```

```
sns.countplot(x=df.type)
plt.show()
```

```
sns.barplot(x='country', y='cases', data=df)
plt.show()
```

```
sns.set(rc={'figure.figsize':(40,20)})
sns.barplot(x='country', y='cases', ci=None, data=df)
plt.xticks(rotation=90)
plt.show()
```

```
sns.set(rc={'figure.figsize':(30,20)})
sns.barplot(x='date', y='cases', ci=None, data=df)
plt.xticks(rotation=90)
plt.show()
```

```
df1 = df[(df['country'] == 'US') | (df['country'] == 'Russia') | (df['country'] == 'Brazil')]
df1.shape
```

```
sns.set(rc={'figure.figsize':(30,10)})
sns.lineplot(x='date', y='cases', hue='country', ci=None, data=df1)
plt.xticks(rotation=90)
plt.show()
```

```
sns.barplot(x='country', y='cases', hue='type', ci=None, data=df1)
plt.show()
```

```
sns.relplot(x='date', y='cases', row='country', 
            height=5,
            aspect=5,
            ci=None, kind='line',data=df1)
plt.xticks(rotation=90)
plt.show()
```
